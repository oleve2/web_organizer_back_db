```python
from datetime import *
import sqlite3
```

```python
conn = sqlite3.connect('./sample_db.db')
```

#### 01) Создание основных таблиц

```python
sql_01_01_posts = """
CREATE TABLE "posts" (
	"id"	INTEGER,
	"title"	TEXT,
	"text"	TEXT,
	"theme"	TEXT,
	"part"	TEXT,
	PRIMARY KEY("id" AUTOINCREMENT)
);
"""

sql_01_02_activ_log = """
CREATE TABLE "activ_log" (
	"id"	INTEGER,
	"activ_name_id"	INTEGER NOT NULL,
	"activ_norm_id"	INTEGER NOT NULL,
	"activ_date"	TEXT NOT NULL,
	"activ_value"	INTEGER NOT NULL,
	PRIMARY KEY("id" AUTOINCREMENT)
);
"""

sql_01_03_activ_names = """
CREATE TABLE "activ_names" (
	"id"	INTEGER,
	"name"	TEXT,
	"date_start"	TEXT,
	"date_end"	TEXT,
	"norm_id"	INTEGER,
	PRIMARY KEY("id" AUTOINCREMENT)
);
"""

sql_01_04_activ_normative = """
CREATE TABLE "activ_normative" (
	"id"	INTEGER,
	"name"	TEXT,
	"norm_period"	TEXT,
	"norm_value"	INTEGER,
	"norm_measure"	TEXT,
	PRIMARY KEY("id" AUTOINCREMENT)
);
"""

sql_01_05_dates = """
CREATE TABLE "dates" (
	"date"	TEXT UNIQUE,
	PRIMARY KEY("date")
);
"""
```

```python
cur1 = conn.cursor()
```

```python
sql_list = [
    sql_01_01_posts,
    sql_01_02_activ_log,
    sql_01_03_activ_names,
    sql_01_04_activ_normative,
    sql_01_05_dates
]
```

```python
for elem in sql_list:
    print(elem)
    cur1.execute(elem)
    print(" ==> done \n -------------------------\n")
```

```python
conn.commit()
```

#### 02) Заполнение таблицы dates

```python
total_dates = []

date_from = date.fromisoformat('2022-01-01')
date_to = date.fromisoformat('2024-01-01')

curr_date = date_from
while (curr_date <= date_to):
    d = ( str(curr_date), ) #{"date": str(curr_date)}
    #print(d)
    total_dates.append(d)
    curr_date += timedelta(days=1)
```

```python
total_dates
```

#### Вставка данных

```python
# вставка данных

def insert_date(conn, dataObj):
    for elem in dataObj:
        #print(elem)
        #input()
        sql = """insert into dates(date) VALUES(?)"""
        cur = conn.cursor()
        cur.execute(sql, elem ) 
    print('done')

#
insert_date(conn, total_dates)
```

```python
conn.commit()
```
